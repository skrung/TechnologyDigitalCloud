## &#49; คณิตศาสตร์สำหรับการเงิน:เงินทบต้นกับดอกเบี้ย

กำหนดให้ เรามีเงินลงทุนตั้งต้น $P$ บาท (Principal) จะนำไปฝากสถาบันการเงินที่ให้ ดอกเบี้ยในแต่ละปี คือ $100r\%$ (Nominal rate) จะได้เงินทบต้น (Compound amount) ที่เกิดจาก $P$ (Principal) บวกกับดอกเบี้ยที่ได้เมื่อครบปี $r P$

ตัวอย่างเช่น เงินต้น $P = 1000$ บาท กับดอกเบี้ยเงินฝาก $0.30\%$ ต่อปี หรือ $r = \frac{3}{1000}$ จะได้เงินเก็บเมื่อครบปี เท่ากับ $1000 + 3 \times 1000 = 1003$ บาท เราเรียกเงินเก็บที่เกินจากเงินต้นว่า ดอกเบี้ยทบต้น

```python
# เงินต้น
P = 1000
OriginalP = P
# ดอกเบี่ยรายปี
r = 0.3/100
# เงินทบต้นเมื่อท้ายปี
S = P + r*P
print("จำนวนเงินทบต้นของหนึ่งปีคือ", S, " ดอกเบี่ยทบต้นคือ ", S-OriginalP)
```

```python
# ปีที่สอง
P = S
# เงินทบต้นเมื่อท้ายปี
S = P + r*P
print("จำนวนเงินทบต้นของสองปีคือ", S, " ดอกเบี่ยทบต้นคือ ", S-OriginalP)
```

```python
# ปีที่สาม
P = S
# เงินทบต้นเมื่อท้ายปี
S = P + r*P
print("จำนวนเงินทบต้นของสามปีคือ", S, " ดอกเบี่ยทบต้นคือ ", S-OriginalP)
```

### โจทย์ 1 เงินทบต้น
กำหนดให้ผู้ลงทุนมีเงินต้น 10000 บาท ต้องการฝากในธนาคารด้วยดอกเบี่ยต่อปีคือ 2\% จงหาเงินทั้งหมดที่ได้หลังปีที่ 2

## &#50; สูตรคณิตศาสตร์สำหรับเงินทบต้นของ $n$ ปี

จากสูตรการคำนวณเงินทบต้น $P$ ปีแรก:
$$P + rP = P(1 + r)^1$$  
สูตรการคำนวณเงินทบต้น $P$ ปีที่สอง:
$$P(1+r)^1 + r P(1+r)^1 = P(1 + r)^2$$  
สูตรการคำนวณเงินทบต้น $P$ ปีที่สาม:
$$P(1+r)^2 + r P(1+r)^2 = P(1 + r)^3$$  

สำหรับปีที่ $n$ $$S = P(1+r)^n$$ เมื่อ $S$ คือเงินทบต้น $P$ คือเงินต้น และ $r$ คืออัตราดอกเบี้ยรายปี


```python
def CompoundAmount(P, r, n):
  ''' CompoundAmount คำนวณเงินทบต้นจากการลงทุน P ด้วยอัตราดอกเบี้ย 100r% ต่อปี จำนวน n ปี
  Input:P = เงินต้น
        r = อัตราดอกเบี่ย 100r%
        n = จำนวนปี
  Output: จำนวนเงินรวมทบต้น
  '''
  return P*(1+r)**n
```

```python
P, r, n = 1000, 3/1000, 4
S = CompoundAmount(P, r, n)
print("จำนวนเงินทบต้นของสามปีคือ", S, " ดอกเบี่ยทบต้นคือ ", S-P)
```

### โจทย์ 2 เงินทบต้น
กำหนดให้ผู้ลงทุนมีเงินต้น 2000 บาท ต้องการฝากในธนาคารด้วยดอกเบี่ยต่อปีคือ 3\% จงหาเงินทั้งหมดที่ได้หลังปีที่ 10

## &#51; สูตรเงินทบต้นของ $n$ ปีที่ให้ดอกเบี้ยก่อนครบปี

พิจารณาว่าสถาบันการเงินให้ดอกเบี้ยทุก ๆ 3 เดือนในหนึ่งปี เรียกว่า ดอกเบี่ยรายไตรมาส ให้ $P$ คือเงินต้น $r$ คืออัตราดอกเบี้ยรายปี $100r\%$ มีการคิดดอกเบี่ยเป็นรายไตรมาส ดังนั้นในหนึ่งปีจะมีการคิดดอกเบี้ยสี่ครั้ง
$$S = P(1+r/4)^4$$

เมื่อคิดเป็น $n$ ปีจะได้สูตรเป็น
$$S = P(1+r/4)^{4 n}$$


```python
def CompoundAmount(P, r, n, m=1):
  ''' CompoundAmount คำนวณเงินทบต้นจากการลงทุน P ด้วยอัตราดอกเบี้ย 100r% ต่อปี จำนวน n ปี
  แต่ละปีมีการคิดดอกเบี้บให้ m ครั้ง
  Input:P = เงินต้น
        r = อัตราดอกเบี่ย 100r%
        n = จำนวนปี
        m = จำนวนครั้งที่คิดคอกเบี้ยให้ใน 1 ปี
  Output: จำนวนเงินรวมทบต้น
  '''
  return P*(1+r/m)**(m*n)
```

```python
# เงินต้น
P = 1000
# อัตราดอกเบี้ยต่อปี
r = 0.3/100
# จำนวนปี
n = 10

# ดอกเบี้ยคิดเป็นรายปี
m = 1
print("จำนวนเงินทบต้น รายปีเท่ากับ", CompoundAmount(P, r, n, m))

# ดอกเบี้ยคิดเป็นรายครึ่งปี
m = 2
print("จำนวนเงินทบต้น รายครึ่งปีเท่ากับ", CompoundAmount(P, r, n, m))

# ดอกเบี้ยคิดเป็นรายไตรมาส
m = 4
print("จำนวนเงินทบต้น รายไตรมาสเท่ากับ", CompoundAmount(P, r, n, m))

# ดอกเบี้ยคิดเป็นรายเดือน
m = 12
print("จำนวนเงินทบต้น รายเดือนเท่ากับ", CompoundAmount(P, r, n, m))

# ดอกเบี้ยคิดเป็นรายสัปดาห์
m = 52
print("จำนวนเงินทบต้น รายสัปดาห์เท่ากับ", CompoundAmount(P, r, n, m))

# ดอกเบี้ยคิดเป็นรายวัน
m = 365
print("จำนวนเงินทบต้น รายวันเท่ากับ", CompoundAmount(P, r, n, m))
```

### โจทย์ 3 เงินทบต้นคิดดอกเบี้ยก่อนครบปี

สมมติเรามีเงิน 3000 บาทที่จะฝากในบัญฃีออมทรัพย์ โดยธนาคารให้ดอกเบี่ย 0.6% ทบรายครึ่งปี จะหายอดเงินในบัญชีนี้หลังผ่านไปเจ็ดปี (สมมติว่าไม่มีการฝากเงินเพิ่มหรือถอนเงินออกจากบัญชี)

## &#52; อัตราดอกเบี้ยที่แท้จริง

ในกรณีที่เงินลงทุน $P$ บาท ฝากในสถาบันการเงินที่ให้ดอกเบี้ยรายปี $6\%$ ดอกเบี้ยคิดแบบรายไตรมาส ในหนึ่งปีจะได้ดอกเบี้ยมากกว่า $6\%$

ดอกเบี้ยที่ได้จากเงินทบต้นในหนึ่งปีได้จาก $$\begin{array}{lcl} S - P & = & \displaystyle P \left(1 + \frac{\frac{6}{100}}{4} \right)^4 - P \\ & = & \displaystyle \left[ (1.015)^4 - 1 \right] P \\ & \approx & 0.06136355062499965 P \end{array}$$

หรือได้ประมาณ $6.14\%$ ของ $P$ เรียก อัตราดอกเบี้ยที่ได้นี้ว่า **อัตราดอกเบี้ยที่แท้จริง (effective rate)** หรืออีกชื่อหนึ่งคือ Annual equivalent rate (AER)

กำหนดให้ $r_e$ คืออัตราดอกเบี้ยที่แท้จริง และ $r$ คืออัตราดอกเบี้ยต่อปี มีการคิดดอกเบี้ย $m$ ครั้งในหนึ่งปี

$$\begin{array}{lcl} P(1 + r_e) & = & \displaystyle P \left(1 + \frac{r}{m} \right)^m \\ 1 + r_e & = & \displaystyle \left( 1 + \frac{r}{m} \right)^m \\ r_e & = & \displaystyle\left(1 + \frac{r}{m} \right)^m - 1 \end{array}$$



```python
# เงินต้น
P = 1
# อัตราดอกเบี้ยต่อปี
r = 6/100
# จำนวนปี
n = 1
# ดอกเบี้ยคิดเป็นรายไตรมาส
m = 4
S = CompoundAmount(P, r, n, m)
er = 100*(S - P)
print("อัตราดอกเบี้ยที่แท้จริงคือ %4.2f%s"%(er,'%'))
```

## &#53; มูลค่าปัจจุบัน

สำหรับเงินลงทุน $P = 100$ บาท ถูกฝากในบัญชีออมทรัพย์ที่มีอัตราดอกเบี้ยรายปีคือ $0.3\%$ เมื่อท้ายปีที่สองจำนวนเงินทบต้นคือ $100(1.003)^2 = 100.6009$

เรากล่าวว่า เงินทบต้น 100.6009 เป็นเงินอนาคตของ 100 ณ เวลาปัจจุบัน

เราเรียกว่า 100 คือมูลค่าปัจจุบันของเงินอนาคตสองปี 100.6009

**สูตรของมูลค่าปัจจุบัน** ของเงินอนาคต $S$ ด้วยอัตราดอกเบี้ยรายปีคือ $$P = S(1 + r)^{-n}$$
